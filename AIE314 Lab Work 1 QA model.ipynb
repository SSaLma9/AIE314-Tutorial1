```python
from docx import Document

def extract_text_from_word(file_path):
    # Load the document
    doc = Document(file_path)

    # Initialize an empty string to hold the text
    full_text = []

    # Iterate through each paragraph in the document
    for para in doc.paragraphs:
        full_text.append(para.text)

    # Join all paragraphs into a single string
    return '\n'.join(full_text)

# Example usage
file_path = 'example.docx'
text = extract_text_from_word(file_path)
print(text)
```

```python
from docx import Document

def extract_first_500_words(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.extend(para.text.split())  # Split paragraph into words

    # Extract first 500 words (or fewer if document is shorter)
    first_500_words = full_text[:500]

    return " ".join(first_500_words)  # Join words into a string

# Example usage
file_path = 'example.docx'
text_500 = extract_first_500_words(file_path)
print(text_500)
```

```python
from docx import Document
import re

def extract_and_clean_first_500_words(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.extend(para.text.split())  # Split paragraph into words

    # Extract first 500 words (or fewer if document is shorter)
    first_500_words = full_text[:500]

    # Remove spaces and split into words using regex
    cleaned_words = re.findall(r'\b[a-zA-Z]+\b', " ".join(first_500_words))

    return cleaned_words  # Return the cleaned words

# Example usage
file_path = 'example.docx'
cleaned_text_500 = extract_and_clean_first_500_words(file_path)
print(cleaned_text_500)
```

```python
import PyPDF2

def extract_text_from_pdf(file_path):
    # Open the PDF file in read-binary mode
    pdf_file = open(file_path, 'rb')

    # Create a PDF file reader object using PdfReader
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)

    # Initialize an empty string to store the text
    pdf_text = ""

    # Loop through each page and extract the text
    for page in range(num_pages):
        page_obj = pdf_reader.pages[page]
        pdf_text += page_obj.extract_text()

    # Close the PDF file
    pdf_file.close()

    # Return the extracted text
    return pdf_text

# Example usage
file_path = 'example.pdf'
text = extract_text_from_pdf(file_path)
print(text)
```

```python
from bs4 import BeautifulSoup

def extract_text_from_html(file_path):
    # Open and read the HTML file
    with open(file_path, "r") as file:
        html_content = file.read()

    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, "html.parser")

    # Extract and print the text
    text = soup.get_text()

    return text

# Example usage
file_path = "example.html"
extracted_text = extract_text_from_html(file_path)
print(extracted_text)
```
